```python
df = pd.read_excel("Technological-Products-Sample-Data.xlsx", header=5)
df.head()
```

```python
df = df.loc[:, ~df.columns.str.contains("Unnamed")]
df.head()
```

```python
import matplotlib.pyplot as plt

df["Brand"].value_counts().plot(kind="bar")

plt.title("Number of Products by Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["Price (USD)"], bins=10)

plt.title("Price Distribution")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Country of Origin"].value_counts().plot(kind="bar")

plt.title("Products by Country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
top10 = df.sort_values(by="Price (USD)", ascending=False).head(10)
top10
```

```python
df["Brand"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(7,7)
)

plt.title("Brand Distribution")
plt.ylabel("")
plt.show()
```
